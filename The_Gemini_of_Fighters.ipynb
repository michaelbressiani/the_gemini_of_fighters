```python
dic = {
  "forca": 90,
  "velocidade": 60,
  "agilidade": 20,
  "defesa": 40,
  "soco": 5,
  "cotovelada": 5,
  "chute": 70,
  "joelhada": 75,
  "defender": 80,
  "golpe_especial": 90,
  "descricao_golpe_especial": "Mandíbula feroz: Uma investida poderosa combinada com uma mordida esmagadora, finalizada por uma chicotada brutal com a cauda."
}

# Função para carregar a imagem da URL e converter para array NumPy
def carregar_imagem_da_url(url):
  resposta = requests.get(url)
  imagem = PILImage.open(BytesIO(resposta.content))
  return np.array(imagem)

# Função para plotar imagens lado a lado

def plot_images_3(images_players, names_image):

  url_image = []

  for i in range(0, 3):
    url_image.append(carregar_imagem_da_url(images_players[i]))

  fig, axes = plt.subplots(1, len(images_players), figsize=(10, 5))

  axes[0].imshow(url_image[0])
  axes[0].axis('off')
  axes[0].set_title(names_image[0])

  axes[1].imshow(url_image[1])
  axes[1].axis('off')
  axes[1].set_title(names_image[1])

  axes[2].imshow(url_image[2])
  axes[2].axis('off')
  axes[2].set_title(names_image[2])

  plt.show()

def plot_images_2(images_players, names_image):

  url_image = []

  for i in range(0, 2):
    url_image.append(carregar_imagem_da_url(images_players[i]))

  fig, axes = plt.subplots(1, len(images_players), figsize=(10, 5))

  axes[0].imshow(url_image[0])
  axes[0].axis('off')
  axes[0].set_title(names_image[0])

  axes[1].imshow(url_image[1])
  axes[1].axis('off')
  axes[1].set_title(names_image[1])

  plt.show()

# Função para plotar imagem isolada

def isolation_plot(image, name_image):

    image_img = carregar_imagem_da_url(image)

    fig, ax = plt.subplots(figsize=(10, 5))

    ax.imshow(image_img)
    ax.axis('off')
    ax.set_title(name_image)
    plt.show()


def print_attributes(dic):
  print("Atributos físicos gerais:", "")
  print("Força:      ", dic["forca"], "*" * int(dic["forca"]/10))
  print("Velocidade: ", dic["velocidade"], "*" * int(dic["velocidade"]/10))
  print("Agilidade:  ", dic["agilidade"], "*" * int(dic["agilidade"]/10))
  print("Defesa:     ", dic["defesa"], "*" * int(dic["defesa"]/10), "")
  print("Ataque e defesa:", "")
  print("Soco:       ", dic["soco"], "*" * int(dic["soco"]/10))
  print("Cotovelada: ", dic["cotovelada"], "*" * int(dic["cotovelada"]/10))
  print("Chute:      ", dic["chute"], "*" * int(dic["chute"]/10))
  print("Joelhada:   ", dic["joelhada"], "*" * int(dic["joelhada"]/10))
  print("Defender:   ", dic["defender"], "*" * int(dic["defender"]/10))
  print("Especial:   ", dic["golpe_especial"], "*" * int(dic["golpe_especial"]/10), "")
  print("O que faz o especial:   ", dic["descricao_golpe_especial"], "")

def array_attributes():
   return [dic["forca"], dic["velocidade"], dic["agilidade"], dic["defesa"]]

def attacks():
  return [dic["soco"], dic["cotovelada"], dic["chute"], dic["joelhada"], dic["defender"], dic["golpe_especial"]]

def proba_moves(dic):
  proba = []
  for a in attacks()[:4]:
    proba.append( ( (dic["forca"] + dic["velocidade"] + a) / 300 ) * random.random() )
  proba.append( ( ( (dic["agilidade"] + dic["defesa"] + dic["defender"]) / 300 ) * 0.5 ) * random.random() )
  proba.append( ( ((dic["forca"] + dic["velocidade"] + dic["agilidade"] + dic["defesa"] ) / 400 ) * 0.5) * random.random() )
  return proba

def score_jogador(moves_player, proba_moves_player):

  proba_player = []
  for i in range(0, len(moves_player)):
    proba_player.append(proba_moves_player[moves_player[i]])

  return np.dot(moves_player, proba_player)

def optm_random_moves_adversary(proba_moves_adversary):

  damege_adversary = []
  moves_adversary_list = []

  for i in range(0, 50000):

    moves_adversary = []
    while len(moves_adversary) < 6:
        novo_numero = random.randint(0, 5)  # Altere os limites conforme necessário
        if not moves_adversary or novo_numero != moves_adversary[-1]:
            moves_adversary.append(novo_numero)

    proba_adversary = []
    for i in range(0, len(moves_adversary)):
        proba_adversary.append(proba_moves_adversary[moves_adversary[i]])

    damege_adversary.append(np.dot(moves_adversary, proba_adversary))
    moves_adversary_list.append(moves_adversary)

  return moves_adversary_list[damege_adversary.index(max(damege_adversary))], damege_adversary[damege_adversary.index(max(damege_adversary))]

def statistics_and_final_result(result, moves_player, moves_adversary, proba_moves_player, proba_moves_adversary):
  if result == -1:
    print('*** Voce perdeu. ***', "\n")
    response = model.generate_content("Reescreva a frase Voce perdeu, de forma descontraida, sem ser ofensivo, e motivando")
    print(response.text, "\n")
  elif result == 1:
    response = model.generate_content("Reescreva a frase Parabéns, voce venceu!, de forma descontraida, sem ser ofensivo, e comemorando")
    print(response.text, "\n")
    print('*** Parabéns, voce venceu! ***', "\n")
  else:
    print('*** Que coincidencia, deu empate! ***', "\n")

  print("Sequencia de golpes escolhidos por voce e score da sua sequencia:", "\n")

  for i in moves_player:
    if i == 0:
      print("Soco")
    elif i == 1:
      print("Cotovelada")
    elif i == 2:
      print("Chute")
    elif i == 3:
      print("Joelhada")
    elif i == 4:
      print("Defender")
    else:
      print("Especial")

  print("Score do Jogador:", round(proba_moves_player, 2), "\n")

  print("Sequencia de golpes escolhidos pelo do Adversário e score da sequencia:", "\n")

  for i in moves_adversary:
    if i == 0:
      print("Soco")
    elif i == 1:
      print("Cotovelada")
    elif i == 2:
      print("Chute")
    elif i == 3:
      print("Joelhada")
    elif i == 4:
      print("Defender")
    else:
      print("Especial")

  print("Score do Adversário:", round(proba_moves_adversary, 2))
```
